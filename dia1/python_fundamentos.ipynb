Regla de oro para definir funciones:

> **Una responsabilidad, una función**

Es decir, si llevas 29 líneas de código en la definición de una función, probablemente no hayas conceptualizado bien su diseño

````python
# Escenario: Necesitamos analizar datos de una acción, calculando su rendimiento
# y algunos indicadores estadísticos básicos

# ❌ MAL EJEMPLO: Una función que hace demasiado
def analizar_accion(precios, nombre_accion):
    """Esta función hace demasiadas cosas diferentes."""
    # Calcula rendimiento
    rendimiento = (precios[-1] - precios[0]) / precios[0] * 100

    # Calcula estadísticas
    media = sum(precios) / len(precios)
    varianza = sum((x - media) ** 2 for x in precios) / len(precios)
    desviacion = varianza ** 0.5

    # Encuentra máximos y mínimos
    maximo = max(precios)
    minimo = min(precios)

    # Genera reporte
    print(f"Análisis de {nombre_accion}")
    print(f"Rendimiento: {rendimiento:.2f}%")
    print(f"Media: {media:.2f}")
    print(f"Desviación Estándar: {desviacion:.2f}")
    print(f"Máximo: {maximo}")
    print(f"Mínimo: {minimo}")

    return rendimiento

# ✅ BUEN EJEMPLO: Funciones con responsabilidad única
def calcular_rendimiento(precios):
    """Calcula el rendimiento porcentual."""
    return (precios[-1] - precios[0]) / precios[0] * 100

def calcular_estadisticas(precios):
    """Calcula estadísticas básicas."""
    media = sum(precios) / len(precios)
    varianza = sum((x - media) ** 2 for x in precios) / len(precios)
    return {
        'media': media,
        'desviacion': varianza ** 0.5,
        'maximo': max(precios),
        'minimo': min(precios)
    }

def generar_reporte(nombre_accion, rendimiento, stats):
    """Genera un reporte formateado."""
    print(f"Análisis de {nombre_accion}")
    print(f"Rendimiento: {rendimiento:.2f}%")
    print(f"Media: {stats['media']:.2f}")
    print(f"Desviación Estándar: {stats['desviacion']:.2f}")
    print(f"Máximo: {stats['maximo']}")
    print(f"Mínimo: {stats['minimo']}")

# Ejemplo de uso de ambos enfoques
precios_ejemplo = [100, 102, 98, 103, 105]
nombre = "AAPL"

# Uso del mal ejemplo
print("Análisis con función monolítica:")
analizar_accion(precios_ejemplo, nombre)

print("\nAnálisis con funciones separadas:")
# Uso del buen ejemplo
rendimiento = calcular_rendimiento(precios_ejemplo)
stats = calcular_estadisticas(precios_ejemplo)
generar_reporte(nombre, rendimiento, stats)
````

# 7. Programación Orientada a Objetos (POO)

La POO permite organizar el código en clases, que a su vez definen objetos con atributos y métodos.

## 7.1 Clase CuentaBancaria
A continuación, un ejemplo que modela una cuenta bancaria sencilla, con atributos y métodos para depositar y retirar dinero.

````python
class CuentaBancaria:
    """Representa una cuenta bancaria sencilla."""

    def __init__(self, titular: str, saldo_inicial: float = 0) -> None:
        """Inicializa la cuenta.

        Args:
            titular (str): Nombre del titular.
            saldo_inicial (float): Saldo inicial.
        """
        self.titular = titular
        self.saldo = saldo_inicial
        self.historial = []

    def depositar(self, cantidad: float) -> None:
        """Agrega una cantidad al saldo.

        Args:
            cantidad (float): Monto a depositar.
        """
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser mayor que 0")
        self.saldo += cantidad
        self.historial.append({"tipo": "depósito", "cantidad": cantidad})

    def retirar(self, cantidad: float) -> None:
        """Resta una cantidad del saldo, si hay fondos suficientes.

        Args:
            cantidad (float): Monto a retirar.
        """
        if cantidad <= 0:
            raise ValueError("La cantidad debe ser mayor que 0")
        if cantidad > self.saldo:
            raise ValueError("Saldo insuficiente")
        self.saldo -= cantidad
        self.historial.append({"tipo": "retiro", "cantidad": cantidad})

    def obtener_estado(self) -> dict:
        """Retorna un resumen de la cuenta.

        Returns:
            dict: Con titular, saldo y número de transacciones.
        """
        return {
            "titular": self.titular,
            "saldo": self.saldo,
            "num_transacciones": len(self.historial)
        }
````
